```python
# jogo de adivinhação
#o susuario vai digitar o numero de 0 a 10 para adivinhar o numero secreto
# o usuario tera 03 tentativas
# o programa devera imprimir uma msg se o palpite é maior ou menos que o numero secreto
#o programa deverá imprimir uma msg de quente, morno ou frio

#Regras
#se a diferença do palpite para o numero secreto for 3- frio, 2- morno e 1- quente


import random

a = random.randint(0, 10)
tentativas_max = 0


while tentativas_max < 3:
    numero = int(input("Digite um número de 0 a 10: "))

    if numero == a:
        print("Está correto!")
        break
    elif numero > a:
        print("Seu palpite é maior que o número secreto.")
    elif numero < a:
        print("Seu palpite é menor que o número secreto.")


    tirar = abs(numero - a)

    if tirar == 1:
        print("Quente!")

    if tirar == 2:
        print("Morno!")

    if tirar >= 3:
        print("Frio!")

    tentativas_max += 1  #

if numero != a:
    print("Você perdeu!")
    print("O número secreto era:", a)





```
