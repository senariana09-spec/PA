```python
#fazer um programa que recebe as notas dos 04 bimestres e imprime a média com
#o resultado de aprovado, recuperação ou reprovado
#regras: utilizar while
#nota maior ou igual a 7 aprovado
#nota igual ou maior que 5 recuperação
#nota menor q 5 reprovado

nota1 = int(input("digite sua primeira nota: "))
nota2 = int(input("digite sua segunda nota: "))
nota3 = int(input("digite sua terceira nota: "))
nota4 = int(input("digite sua quarta nota: "))

media = ""

media = nota1 + nota2 + nota3 + nota4/ 4

while media >= 7:
 print ("aprovado")

while media >= 5:
 print ("recuperação")

while media < 4:
 print("reprovado")

media +=1

```
