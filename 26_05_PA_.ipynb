```python
#calculadora simples com as 4 operaçôes
#com menu de opçôes
#1 somar
#2 subtrair
#3 multiplicar
#4 divisão
#5 sair



print("menu de opçôes")
print("opção 1: soma")
print("opção 2: subtração")
print("opção 3: multiplicação: ")
print("opção 4: divisão")
print("opção 5: sair")

opcao = int(input("digite sua opção:"))



if opcao == 1:
  a = int(input("digite o primeiro numero:  "))
  b = int(input("digite o segundo numero: "))
  print(f"o resultado da soma é {a + b}")

elif opcao == 2:
  c = int(input("digite o primeiro numero: "))
  d = int(input("digite o segundo numero: "))
  print(f"o resultado da subtração é {c - d}")

if opcao == 3:
  e = int(input("digite o primeiro numero: "))
  f = int(input("digite o segundo numero: "))
  print(f"o resultado da sua multiplicação é {e * f}")

elif opcao == 4:
  g = int(input("digite o primeiro numero: "))
  h = int(input("digite o segundo numero: "))
  print(f"o resultado da sua divisão é {g / h}")

if opcao == 5:
  print("vc saiu da calculadora")

```

```python
#jogo inteligente
# 1 par ou impar
# 2 positivo ou negativo
# 3 digitar dois numeros e a maquina falar qual o maior
# 4 sair

print("bem vindo ao jogo inteligente")
print("opção 1: descubra se o seu numero é par ou impar ")
print("opção 2: descura se o seu número é positivo ou negativo")
print("opção 3: descubra qual número é maior")
print("opção 4: sair do jogo")

opcao = int(input("digite sua opção:  "))

if opcao == 1:
  a = int(input(f"digite o numero: "))
  if a % 2 == 0:
   print("seu numero é par ")
  else:
   print("seu numero é impar")

if opcao == 2:
  b = int(input(f"digite seu numero: "))
  if b > 0:
   print("seu numero é positivo")
  elif b < 0:
    print("seu numero é negativo")

elif opcao == 3:
  c = int(input("digite o primeiro numero: "))
  d = int(input("digite o segundo numero:  "))
  if c > d:
    print(f"o numero {c} é maior que {d}")
  elif c < d:
    print(f"o numero {c} é menor que {d}")

elif opcao == 4:
  print("vc saiu do jogo, até a proxima")
```
